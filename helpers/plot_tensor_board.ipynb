```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df_reward = pd.read_csv('data/run-logs_1-metrics-tag-average reward.csv')
df_step = pd.read_csv('data/run-logs_1-metrics-tag-average steps.csv')
```

```python
df_reward.head()
```

```python
df_step.head()
```

```python
df_r = df_reward[df_reward['Step'] <= 800_000]
plt.plot(df_r['Step'],df_r['Value'])
plt.ylabel('Average Return')
plt.xlabel('Number of Training Steps')
plt.title(f'Training Performance of DQN')
plt.show()
```

```python
df_s  = df_step[df_step['Step'] <= 800_000]
plt.plot(df_s['Step'],df_s['Value'], color='orange')
plt.ylabel('Average Steps')
plt.xlabel('Number of Training Steps')
plt.title(f'Training Performance of DQN')
plt.show()
```
